```python
import pandas as pd
from data_ingestion import load_dataset
```

```python
df = load_dataset()
```

```python
drop_columns = [
    "CustomerID",
    "Count",
    "Lat Long",
    "Churn Score",
    "Churn Reason"
]

df.drop(
    columns=drop_columns,
    inplace=True
)
```

```python
df.info()
```

```python
df["Total Charges"] = pd.to_numeric(
    df["Total Charges"],
    errors="coerce"
)
```

```python
df.isnull().sum()
```

```python
df["Total Charges"].fillna(
    df["Total Charges"].median(),
    inplace=True
)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["LongTermCustomer"] = (
    df["Tenure Months"] > 24
).astype(int)
```

```python
services = [
    "Online Security",
    "Online Backup",
    "Device Protection",
    "Tech Support",
    "Streaming TV",
    "Streaming Movies"
]

df["ServicesCount"] = (
    (df[services] == "Yes")
    .sum(axis=1)
)
```

```python
df["AvgRevenue"] = (
    df["Total Charges"]
    /
    (df["Tenure Months"] + 1)
)
```
